## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("Customer_support_data.csv")

# Display dataset info
print("\nDataset Information:")
df.info()

# Display first few rows
print("\nFirst 5 Rows:")
print(df.head())

# Check for missing values
print("\nMissing Values Count:")
print(df.isnull().sum())

# Fill missing categorical values with mode (most frequent value)
df["Customer_City"].fillna("Unknown", inplace=True)
df["Product_category"].fillna(df["Product_category"].mode()[0], inplace=True)

# Fill missing text-based column with 'No Remarks'
df["Customer Remarks"].fillna("No Remarks", inplace=True)

# Fill missing order ID with 'Unknown'
df["Order_id"].fillna("Unknown", inplace=True)

# Fill missing order date with mode (most frequent date)
df["order_date_time"].fillna(df["order_date_time"].mode()[0], inplace=True)

# Fill missing numerical values with median
df["Item_price"].fillna(df["Item_price"].median(), inplace=True)

# Drop 'connected_handling_time' due to excessive missing values
df.drop(columns=['connected_handling_time'], inplace=True)

# Verify missing values after handling
print("\nMissing Values After Handling:")
print(df.isnull().sum())

# Check for duplicate rows
print(f"\nDuplicate Rows: {df.duplicated().sum()}")

# Drop duplicate records
df = df.drop_duplicates()

# **Fix Date Parsing Issue**
date_columns = ["order_date_time", "Issue_reported at", "issue_responded", "Survey_response_Date"]

for col in date_columns:
    try:
        df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)  # Try DD/MM/YYYY format first
    except Exception:
        df[col] = pd.to_datetime(df[col], errors='coerce')  # Fall back to automatic parsing

# Verify date conversion
print("\nConverted Date Columns:")
print(df[date_columns].dtypes)

# **Ensure all date conversions worked before using .dt accessor**
if df["order_date_time"].dtype == 'datetime64[ns]':
    # Extract new time-based features
    df["order_year"] = df["order_date_time"].dt.year
    df["order_month"] = df["order_date_time"].dt.month
    df["order_day"] = df["order_date_time"].dt.day
    df["order_weekday"] = df["order_date_time"].dt.dayofweek  # Monday=0, Sunday=6

    # Compute response time in minutes
    df["response_time"] = (df["issue_responded"] - df["Issue_reported at"]).dt.total_seconds() / 60
else:
    print("\nError: `order_date_time` is not in datetime format. Check dataset.")

# Create a handling efficiency metric
df["handling_efficiency"] = df["response_time"] / (df["order_month"] + 1)

# One-hot encode categorical variables
df = pd.get_dummies(df, columns=['channel_name', 'category', 'Sub-category', 'Product_category', 'Tenure Bucket', 'Agent Shift'], drop_first=True)

# Convert CSAT Score to integer (Ordinal Encoding)
df["CSAT Score"] = df["CSAT Score"].astype(int)

# Identify numerical columns
num_cols = ['Item_price', 'response_time']

# Remove outliers using Interquartile Range (IQR)
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Display final dataset info
df.info()

# Save cleaned dataset for further analysis
df.to_csv("cleaned_dataset.csv", index=False)
````

### What all manipulations have you done and insights you found?

* The dataset underwent a comprehensive data wrangling process to ensure it was clean and analysis-ready. Initially, missing values were addressed by filling categorical variables such as Customer_City and Product_category with their most frequent values (mode), while Customer Remarks and Order_id were replaced with "Unknown" where necessary. Numerical variables like Item_price were imputed using the median, ensuring minimal impact on the data distribution. The column connected_handling_time was dropped due to excessive missing values (~99% missing).

* Duplicate records were identified and removed to maintain data integrity. Date columns (order_date_time, Issue_reported at, issue_responded, and Survey_response_Date) were converted into datetime format after handling inconsistencies in date formats. From these, new time-based features such as order_year, order_month, order_day, and order_weekday were extracted. A new feature, response_time, was also calculated by finding the difference between issue_responded and Issue_reported at, which helps in analyzing customer service efficiency.

* To ensure data quality, outliers were detected and removed using the Interquartile Range (IQR) method for numerical variables like Item_price and response_time. This prevented extreme values from skewing the analysis. Additionally, categorical variables (channel_name, category, Sub-category, etc.) were one-hot encoded to make them suitable for machine learning models, and CSAT Score was converted into an integer format.

 Insights Gained:
* Customer Satisfaction Trends: The majority of CSAT Scores were 4 and 5, indicating an overall high satisfaction rate, but further analysis is needed to identify cases of dissatisfaction.
* Service Response Time Impact: Customers with longer response times tended to have lower CSAT scores, indicating that reducing resolution time could improve satisfaction.
* High Missing Values in Certain Columns: Features like Customer Remarks and connected_handling_time had large amounts of missing data, suggesting they may not be consistently recorded or relevant for predictive modeling.
* Variation in Order Processing Time: Some categories had significantly higher response times, possibly due to complex issues, requiring deeper analysis into category-wise agent efficiency.

````python
# Chart - 12 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Agent Shift' is one-hot encoded in your dataset.
# First, reverse the one-hot encoding if needed.

# Create a mapping from one-hot columns to the shift label
shift_columns = ['Agent Shift_Morning', 'Agent Shift_Evening', 'Agent Shift_Night', 'Agent Shift_Split']
shift_labels = ['Morning', 'Evening', 'Night', 'Split']

# Create a new column 'Agent_Shift' from one-hot columns
def get_shift(row):
    for col, label in zip(shift_columns, shift_labels):
        if row.get(col, 0) == 1:
            return label
    return 'Unknown'

df['Agent_Shift'] = df.apply(get_shift, axis=1)

# Plot CSAT Score by Agent Shift
plt.figure(figsize=(10, 6))
sns.boxplot(x='Agent_Shift', y='CSAT Score', data=df, palette='Set2')

plt.title('CSAT Score Distribution by Agent Shift')
plt.xlabel('Agent Shift')
plt.ylabel('CSAT Score')
plt.grid(True)
plt.tight_layout()
plt.show()
````
