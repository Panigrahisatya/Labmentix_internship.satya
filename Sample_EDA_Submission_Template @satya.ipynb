## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Load the merged dataset
merged_data = pd.read_csv("merged_dataset.csv")

# 1. Drop irrelevant or highly missing columns
merged_data.drop(columns=['seasons', 'age_certification'], inplace=True)  # Too many missing values

# 2. Fill missing values
merged_data['description'].fillna("No description available", inplace=True)  
merged_data['imdb_score'].fillna(merged_data['imdb_score'].median(), inplace=True)
merged_data['imdb_votes'].fillna(merged_data['imdb_votes'].median(), inplace=True)
merged_data['tmdb_score'].fillna(merged_data['tmdb_score'].median(), inplace=True)
merged_data['tmdb_popularity'].fillna(merged_data['tmdb_popularity'].median(), inplace=True)
merged_data['character'].fillna("Unknown", inplace=True)

# 3. Convert categorical variables
merged_data['genres'] = merged_data['genres'].apply(lambda x: eval(x) if isinstance(x, str) else x)  # Convert from string to list
merged_data['production_countries'] = merged_data['production_countries'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# 4. Handle outliers (Example: Runtime)
merged_data = merged_data[merged_data['runtime'] < 300]  # Remove extreme runtime values

# 5. Convert lists to strings for easier processing
merged_data['genres'] = merged_data['genres'].apply(lambda x: ', '.join(x) if isinstance(x, list) else "Unknown")
merged_data['production_countries'] = merged_data['production_countries'].apply(lambda x: ', '.join(x) if isinstance(x, list) else "Unknown")

# 6. Convert release_year to integer
merged_data['release_year'] = merged_data['release_year'].astype(int)

# 7. Create a new feature: Decade
merged_data['decade'] = (merged_data['release_year'] // 10) * 10

# 8. Drop duplicates (if any)
merged_data.drop_duplicates(inplace=True)

# 9. Reset index after cleaning
merged_data.reset_index(drop=True, inplace=True)

# Final Check
print(merged_data.info())
print(merged_data.head())

# Save the cleaned dataset
merged_data.to_csv("cleaned_dataset.csv", index=False)
````

### What all manipulations have you done and insights you found?

Manipulations Performed on the Dataset

1. Removed Duplicate Rows – Ensured no redundant data.
2. Handled Missing Values –

* Filled runtime & imdb_score using median.
* Filled age_certification using mode.
* Replaced seasons NaN values with 0 (assuming movies have 0 seasons).
* Replaced tmdb_score missing values with median.

3. Converted Data Types –

* release_year and seasons converted to integers.

4. Parsed List-Like Columns –

* Converted genres and production_countries from string representations to actual lists.

5. One-Hot Encoded Categorical Features –

* Applied one-hot encoding to type (Movie/Show) and age_certification.

6.Created Genre & Country Features –

* Transformed genres and production_countries into binary columns indicating presence.
* Saved the Cleaned Dataset for further analysis.
